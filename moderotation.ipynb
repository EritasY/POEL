```python
import rebound
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from subprocess import call
%matplotlib inline
plt.style.use('paper.mplstyle')
```

```python
def write_png(params):
    i, t, dpi = params
    w1 = 0.2
    e1 = 0.5*0.4

    w2 = 0.3
    e2 = 0.25

    w3 = 0
    e3 = 0
    s=0.87

    fig, axarr = plt.subplots(figsize=(12,6), ncols=2, width_ratios=[1.2, 1])

    #------------------------------------ ax1 ------------------------------------
    ax = axarr[0]
    e2ctrx = e1*np.cos(w1*t)
    e2ctry = e1*np.sin(w1*t)

    ax.arrow(0, 0, s*e2ctrx, s*e2ctry, head_width=0.01, linewidth=4, head_length=0.02, fc='b', ec='b', length_includes_head=True)
    circle = plt.Circle((0, 0), e1, color='b', fill=False, lw=3)
    ax.add_patch(circle)

    e3ctrx = e1*np.cos(w1*t) + e2*np.cos(w2*t)
    e3ctry = e1*np.sin(w1*t) + e2*np.sin(w2*t)

    ax.arrow(e2ctrx, e2ctry, s*(e3ctrx-e2ctrx), s*(e3ctry-e2ctry), head_width=0.01, linewidth=4, head_length=0.02, fc='orange', ec='orange', length_includes_head=True)
    circle = plt.Circle((e2ctrx, e2ctry), e2, color='orange', fill=False, lw=3)
    ax.add_patch(circle)

    e4ctrx = e1*np.cos(w1*t) + e2*np.cos(w2*t) + e3*np.cos(w3*t)
    e4ctry = e1*np.sin(w1*t) + e2*np.sin(w2*t) + e3*np.sin(w3*t)

    ax.arrow(e3ctrx, e3ctry, s*(e4ctrx-e3ctrx), s*(e4ctry-e3ctry), head_width=0.01, linewidth=4, head_length=0.02, fc='k', ec='k', length_includes_head=True)
    circle = plt.Circle((e3ctrx, e3ctry), e3, color='k', fill=False, lw=3)
    ax.add_patch(circle)

    ax.arrow(0, 0, 0.95*e4ctrx, 0.95*e4ctry, head_width=0.01, linewidth=4, head_length=0.02, fc='k', ec='k', length_includes_head=True)

    ax.set_aspect('equal')
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_xlim([-0.6, 0.6])
    ax.set_ylim([-0.6, 0.6])


    #------------------------------------ ax1 ------------------------------------
    ax = axarr[1]
    time = np.linspace(0,t,100)
    e4ctrx = e1*np.cos(w1*time) + e2*np.cos(w2*time) + e3*np.cos(w3*time)
    e4ctry = e1*np.sin(w1*time) + e2*np.sin(w2*time) + e3*np.sin(w3*time)
    e = np.sqrt((e4ctrx**2 + e4ctry**2))
    ax.plot(time,e,color='k')
    ax.set_xlim(0,120)
    ax.set_ylim(0,0.6)
    ax.spines[['right', 'top']].set_visible(False)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_xlabel("t",fontsize=26)
    ax.set_ylabel(r"$e_{12}$",fontsize=26)
    fig.savefig('tmp/{0:0=5d}.png'.format(i), transparent=True, dpi=dpi, bbox_inches='tight')
    plt.close(fig)  
```

```python
i, t, dpi = 0, 60, 300.4
w1 = 0.2
e1 = 0.5*0.4

w2 = 0.3
e2 = 0.25

w3 = 0
e3 = 0
s=0.87

fig, axarr = plt.subplots(figsize=(12,6), ncols=2, width_ratios=[1.2, 1])

#------------------------------------ ax1 ------------------------------------
ax = axarr[0]
e2ctrx = e1*np.cos(w1*t)
e2ctry = e1*np.sin(w1*t)

ax.arrow(0, 0, s*e2ctrx, s*e2ctry, head_width=0.01, linewidth=4, head_length=0.02, fc='b', ec='b', length_includes_head=True)
circle = plt.Circle((0, 0), e1, color='b', fill=False, lw=3)
ax.add_patch(circle)

e3ctrx = e1*np.cos(w1*t) + e2*np.cos(w2*t)
e3ctry = e1*np.sin(w1*t) + e2*np.sin(w2*t)

ax.arrow(e2ctrx, e2ctry, s*(e3ctrx-e2ctrx), s*(e3ctry-e2ctry), head_width=0.01, linewidth=4, head_length=0.02, fc='orange', ec='orange', length_includes_head=True)
circle = plt.Circle((e2ctrx, e2ctry), e2, color='orange', fill=False, lw=3)
ax.add_patch(circle)

e4ctrx = e1*np.cos(w1*t) + e2*np.cos(w2*t) + e3*np.cos(w3*t)
e4ctry = e1*np.sin(w1*t) + e2*np.sin(w2*t) + e3*np.sin(w3*t)

ax.arrow(e3ctrx, e3ctry, s*(e4ctrx-e3ctrx), s*(e4ctry-e3ctry), head_width=0.01, linewidth=4, head_length=0.02, fc='k', ec='k', length_includes_head=True)
circle = plt.Circle((e3ctrx, e3ctry), e3, color='k', fill=False, lw=3)
ax.add_patch(circle)

ax.arrow(0, 0, 0.95*e4ctrx, 0.95*e4ctry, head_width=0.01, linewidth=4, head_length=0.02, fc='k', ec='k', length_includes_head=True)

ax.set_aspect('equal')
ax.set_xticks([])
ax.set_yticks([])
ax.set_xlim([-0.6, 0.6])
ax.set_ylim([-0.6, 0.6])


#------------------------------------ ax1 ------------------------------------
ax = axarr[1]
time = np.linspace(0,t,100)
e4ctrx = e1*np.cos(w1*time) + e2*np.cos(w2*time) + e3*np.cos(w3*time)
e4ctry = e1*np.sin(w1*time) + e2*np.sin(w2*time) + e3*np.sin(w3*time)
e = np.sqrt((e4ctrx**2 + e4ctry**2))
ax.plot(time,e,color='k')
ax.set_xlim(0,120)
ax.set_ylim(0,0.6)
ax.spines[['right', 'top']].set_visible(False)
ax.set_xticks([])
ax.set_yticks([])
ax.set_xlabel("t",fontsize=26)
ax.set_ylabel(r"$e_{12}$",fontsize=26)

fig.savefig('tmp/00000.png'.format(i), transparent=True, dpi=dpi, bbox_inches='tight')   
```

```python
%%time
dpi = 300.4
Lmovie = 10 # sec
fps = 30
t=120
times = np.linspace(0, t, int(Lmovie*fps))

call("rm -f /*.png", shell=True)
pool = rebound.InterruptiblePool()
params = [(i, time, dpi) for i, time in enumerate(times)]
res = pool.map(write_png, params)
```
